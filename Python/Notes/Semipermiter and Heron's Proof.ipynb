## Formulation

---

Heron's formula states that the area of a triangle whose sides have lengths a, b, and c is

\begin{equation}
{\displaystyle A={\sqrt {s(s-a)(s-b)(s-c)}}} 
\end{equation}
where s is the semiperimeter of the triangle; that is,


\begin{equation}
{\displaystyle s={\frac {a+b+c}{2}}}
\end{equation}

Heron's formula can also be written as

\begin{equation}
 A={\frac {1}{4}}{\sqrt {(a+b+c)(-a+b+c)(a-b+c)(a+b-c)}}
\end{equation}

\begin{equation}
A=\frac{1}{4}\sqrt{(a+b+c)(-a+b+c)(a-b+c)(a+b-c)}
\end{equation}

\begin{equation}
A={\frac {1}{4}}{\sqrt {2(a^{2}b^{2}+a^{2}c^{2}+b^{2}c^{2})-(a^{4}+b^{4}+c^{4})}}
\end{equation}

\begin{equation}
 A=\frac{1}{4}\sqrt{2(a^2 b^2+a^2c^2+b^2c^2)-(a^4+b^4+c^4)}
\end{equation}

\begin{equation}
A={\frac {1}{4}}{\sqrt {(a^{2}+b^{2}+c^{2})^{2}-2(a^{4}+b^{4}+c^{4})}}
\end{equation}
 

\begin{equation}
A=\frac{1}{4}\sqrt{(a^2+b^2+c^2)^2-2(a^4+b^4+c^4)}
\end{equation}
 

\begin{equation}
A={\frac {1}{4}}{\sqrt {4a^{2}b^{2}-(a^{2}+b^{2}-c^{2})^{2}}}
\end{equation}

\begin{equation}
A=\frac{1}{4}\sqrt{4a^2b^2-(a^2+b^2-c^2)^2}
\end{equation}
 
 

 

 


### Algebraic proof using the Pythagorean theorem[edit]

Triangle with altitude h cutting base c into d + (c − d). By the Pythagorean theorem we have b2 = h2 + d2 and a2 = h2 + (c − d)2. Subtracting these yields a2 − b2 = c2 − 2cd. This equation allows us to express d in terms of the sides of the triangle:

\begin{equation}
d={\frac {-a^{2}+b^{2}+c^{2}}{2c}} 
\end{equation}

\begin{equation}
d=\frac{-a^2+b^2+c^2}{2c}
\end{equation}



For the height of the triangle we have that h2 = b2 − d2. By replacing d with the formula given above and applying the difference of squares identity we get


 {\begin{aligned}h^{2}&=b^{2}-\left({\frac {-a^{2}+b^{2}+c^{2}}{2c}}\right)^{2}\\&={\frac {(2bc-a^{2}+b^{2}+c^{2})(2bc+a^{2}-b^{2}-c^{2})}{4c^{2}}}\\&={\frac {((b+c)^{2}-a^{2})(a^{2}-(b-c)^{2})}{4c^{2}}}\\&={\frac {(b+c-a)(b+c+a)(a+b-c)(a-b+c)}{4c^{2}}}\\&={\frac {2(s-a)\cdot 2s\cdot 2(s-c)\cdot 2(s-b)}{4c^{2}}}\\&={\frac {4s(s-a)(s-b)(s-c)}{c^{2}}}\end{aligned}} 
\begin{align}
h^2 & = b^2-\left(\frac{-a^2+b^2+c^2}{2c}\right)^2\\
& = \frac{(2bc-a^2+b^2+c^2)(2bc+a^2-b^2-c^2)}{4c^2}\\
& = \frac{((b+c)^2-a^2)(a^2-(b-c)^2)}{4c^2}\\
& = \frac{(b+c-a)(b+c+a)(a+b-c)(a-b+c)}{4c^2}\\
& = \frac{2(s-a)\cdot 2s\cdot 2(s-c)\cdot 2(s-b)}{4c^2}\\
& = \frac{4s(s-a)(s-b)(s-c)}{c^2}
\end{align}



We now apply this result to the formula that calculates the area of a triangle from its height:

 {\begin{aligned}A&={\frac {ch}{2}}\\&={\sqrt {{\frac {c^{2}}{4}}\cdot {\frac {4s(s-a)(s-b)(s-c)}{c^{2}}}}}\\&={\sqrt {s(s-a)(s-b)(s-c)}}\end{aligned}} 
\begin{align}
A & = \frac{ch}{2}\\
& = \sqrt{\frac{c^2}{4}\cdot \frac{4s(s-a)(s-b)(s-c)}{c^2}}\\
& = \sqrt{s(s-a)(s-b)(s-c)}
\end{align}
